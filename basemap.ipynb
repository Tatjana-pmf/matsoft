#Basemap
[Basemap](http://matplotlib.org/basemap/) je dio Matplotlib biblioteke s funkcijama i metodama za pravljenje geografskih karata. Korištenje Basemapa ilustrirat ćemo na nekim primjerima.

```python
#osnovne biblioteke koje ćemo koristiti
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
```

##Crtanje najkraće udaljenosti

```python
fig = plt.figure(figsize=(14,10))
ax = fig.add_axes([0.1,0.1,0.9,0.9])
# Mercatorova projekcija
m = Basemap(projection='merc',llcrnrlat=-80,urcrnrlat=80,\
            llcrnrlon=-180,urcrnrlon=180,lat_ts=20,resolution='c')
# geografske koordinate Zagreba
zaglat = 45.48; zaglon = 15.58
# geografske koordinate Pekinga
pelat = 39.55; pelon = 116.25

m.drawgreatcircle(zaglon,zaglat,pelon,pelat,linewidth=2,color='b')
m.drawcoastlines()
m.fillcontinents(color='coral',lake_color='aqua')
m.drawmapboundary(fill_color='aqua')
m.drawparallels(np.arange(-90.,91.,30.),labels=[1,1,0,1])
m.drawmeridians(np.arange(-180.,181.,60.),labels=[1,1,0,1])
zx,zy=m(zaglon,zaglat)
px,py=m(pelon,pelat)
ax.text(zx,zy,'Zagreb',fontsize=14,fontweight='bold',ha='center',va='center',color='b')
ax.text(px,py,'Peking',fontsize=14,fontweight='bold',ha='center',va='center',color='r')
ax.set_title('Zagreb - Peking');
```

## Dan i noć

```python
from datetime import datetime
# Millerova projekcija
fig = plt.figure(figsize=(8,8))
ax = fig.add_axes([0,0,1,1])
map = Basemap(projection='mill',lon_0=90)

map.drawcoastlines()
map.drawparallels(np.arange(-90,90,30),labels=[1,0,0,0])
map.drawmeridians(np.arange(map.lonmin,map.lonmax+30,60),labels=[0,0,0,1])

map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral',lake_color='aqua')

date = datetime.utcnow()
CS=map.nightshade(date)
ax.set_title(u'Mapa dana/noći za %s (UTC)' % date.strftime("%d %b %Y %H:%M:%S"));
```

##Crtanje SST (sea surface temperature)
Podaci se preuzimaju sa stranice NOAA (National oceanic and atmospheric administration). Podaci su u netCDF formatu, koji je u verziji 4 u biti podskup HDF formata (Hierarchical Data Format). Opis formata možete naći [ovdje](http://en.wikipedia.org/wiki/HDF5).
Podaci koje uzimamo su _Optimally Interpolated V2 SST AVHRR_ (http://www.ncdc.noaa.gov/thredds/oisst-catalog.html)

U nastavku ćemo koristiti funkciju `squeeze` iz pakata `numpy`. `squeeze` odstranjuje jednodimenzionalne elemente iz niza.

```python
x = np.array([[[0], [1], [2]]])
print(x.shape)
print(np.squeeze(x).shape)
print(np.squeeze(x))
```

```python
from netCDF4 import Dataset, date2index
date = datetime(2013,12,15,0) 

dataset = Dataset('http://www.ncdc.noaa.gov/thredds/dodsC/OISST-V2-AVHRR_agg')
timevar = dataset.variables['time']
timeindex = date2index(date,timevar) # nalazimo vremenski indeks za odabrani datum
#čitanje podataka
sst = dataset.variables['sst'][timeindex,:].squeeze()

lats = dataset.variables['lat'][:]
lons = dataset.variables['lon'][:]
lons, lats = np.meshgrid(lons,lats)

fig = plt.figure(figsize=(8,8))
ax = fig.add_axes([0.05,0.05,0.9,0.9])

m = Basemap(projection='kav7',lon_0=0,resolution=None)

m.drawmapboundary(fill_color='0.3')

im = m.pcolormesh(lons,lats,sst,shading='flat',cmap=plt.cm.jet,latlon=True)

m.drawparallels(np.arange(-90.,99.,30.))
m.drawmeridians(np.arange(-180.,180.,60.))

cb = m.colorbar(im,"bottom", size="5%", pad="2%")

ax.set_title('SST analiza za %s'%date);
```

###Priprema

Podaci se dohvaćaju sa stranice [earthquake.usgs.gov](http://earthquake.usgs.gov/). Učitavamo datoteku u kojoj su dani podaci o zemljotresima u zadnjih tjedan dana. Datoteka se ažurira svakih 5 minuta.

```python
#dohvaćanje podataka

# http://earthquake.usgs.gov/earthquakes/feed/v1.0/csv.php
r = get("http://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_week.csv")
```

```python
#Parsiranje

lat, lon, mag = [], [], []
reader = DictReader(r.text.splitlines())
for redak in reader:
     lon.append(float(redak['longitude']))
     lat.append(float(redak['latitude']))
     mag.append(float(redak['mag']))  
```

```python
#Normalizacija magnituda
norm = Normalize()
mag_norms = norm(np.array(mag))
z = (mag_norms * 10.0)**2.0
```

###Crtanje

```python
fig = figure(figsize=(14,10))


m = Basemap(projection='moll', lat_0=0, lon_0=0, resolution='l', area_thresh=1000.0)

m.drawcoastlines()
m.fillcontinents(color='coral',lake_color='aqua')
m.drawmapboundary(fill_color='aqua')
m.drawmeridians(np.arange(0, 360, 30))
m.drawparallels(np.arange(-90, 90, 30))
cmap = get_cmap('jet')
x,y = m(lon, lat)
sc = m.scatter(x, y, s=z, cmap=cmap, c=mag)
plt.colorbar(sc, orientation='horizontal')

plt.title("Potresi");
```
